```python
# Creating a mock dataset with the given schema for visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'id': range(22706),
    'inning': [1, 2] * 11353,
    'over': [i % 20 + 1 for i in range(22706)],
    'ball': [i % 6 + 1 for i in range(22706)],
    'batsman': [f'Batsman_{i % 50}' for i in range(22706)],
    'bowler': [f'Bowler_{i % 30}' for i in range(22706)],
    'batsman_runs': [i % 7 for i in range(22706)],
    'total_runs': [i % 8 for i in range(22706)],
    'is_wicket': [0 if i % 21 else 1 for i in range(22706)],
    'dismissal_kind': [None if i % 21 else 'Caught' for i in range(22706)],
    'player_dismissed': [None if i % 21 else f'Batsman_{i % 50}' for i in range(22706)],
    'batting_team': [f'Team_{i % 10}' for i in range(22706)],
    'bowling_team': [f'Team_{(i+1) % 10}' for i in range(22578)] + [None] * 128,
}
df = pd.DataFrame(data)
```

```python


df = pd.DataFrame(data)

# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Numerical Feature Correlations")
plt.show()
```

```python
#Pie Chart
wicket_data = df['is_wicket'].value_counts()
labels = ['No Wicket', 'Wicket']
colors = ['skyblue', 'salmon']
plt.figure(figsize=(8, 8))
plt.pie(wicket_data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=[0, 0.1])
plt.title("Proportion of Balls Resulting in Wickets")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='over', y='total_runs', hue='inning', palette="cool", alpha=0.7)
plt.title("Scatter Plot of Runs Scored per Over", fontsize=14)
plt.xlabel("Over", fontsize=12)
plt.ylabel("Total Runs", fontsize=12)
plt.legend(title="Inning", loc="upper right")
plt.grid(alpha=0.3)
plt.show()
```

```python

wickets_per_over = df.groupby('over')['is_wicket'].sum()

# Plotting the line graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=wickets_per_over.index, y=wickets_per_over.values, marker='o', color='r', linewidth=2)
plt.title("Wickets Taken per Over", fontsize=14)
plt.xlabel("Over", fontsize=12)
plt.ylabel("Wickets", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
